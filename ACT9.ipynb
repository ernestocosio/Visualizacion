```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import statsmodels.api as sm
```

```python
%ls datasets
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
df.columns
```

```python
modelA = sm.OLS.from_formula('cadera~peso', data = df)
resultA = modelA.fit()
y = resultA.resid
y.median()
```

```python
vars = ['folio', 'intp', 'entidad', 'desc_ent', 'sexo', 'edad', 'meses', 'peso',
       'ropa', 'talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol',
       'diastol', 'hpresion', 'tbrazo', 'htension', 'PrimaryLast', 'code_upm',
       'est_dis', 'est_urb', 'est_marg', 'pondef', 'est_var']
df = df[vars].dropna()
df
```

```python
df.corr()
```

```python
modelB = sm.OLS.from_formula('diastol~sistol', data = df)
resultB = modelB.fit()
resultB.summary()
```

```python
modelB = sm.OLS.from_formula('diastol~sistol + edad + peso + cintura', data = df)
resultB = modelB.fit()
resultB.summary()
```
