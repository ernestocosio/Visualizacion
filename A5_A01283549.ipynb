```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls datasets
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df
```

```python
sns.boxplot(data=df.loc[:,['peso']])
```

```python
tabla = pd.pivot_table(df,
                       index = ['desc_ent'],
                       columns = ['sexo'],
                       values = ['edad'],
                       aggfunc = {'edad':['median']}
                      )
tabla
```

```python
tabla.columns
```

```python
boxplot = tabla.boxplot(column=[('edad', 'median', 1)])
```

```python
tablados = pd.pivot_table(df,
                    index = ['desc_ent'],
                       columns = ['sexo'],
                       values = ['cadera'],
                       aggfunc = {'cadera':['median']}
                      )
tablados  
```

```python
line_plot = tablados[('cadera', 'median',2)].plot(kind='line')
```

```python
tablatres = pd.pivot_table(df,
                           index = ['desc_ent'],
                       columns = ['sexo'],
                       values = ['ropa'],
                       aggfunc = {'ropa':['sum']}
                      )
tablatres
                          
                           
```

```python
by_ropa = tablatres.sort_values(('ropa','sum',1),ascending=False)
top10 = by_ropa.head(10)
top10
```

```python
top10[('ropa','sum',1)].plot(kind='barh',
                             figsize = (20,10),
                             color = 'y',
                             alpha = 0.5)
```
